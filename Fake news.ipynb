```python
 import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake = pd.read_csv('Fake.csv')
data_true = pd.read_csv('True.csv')
```

```python
#Assigning classes to the dataset

data_fake["class"] = 0
data_true["class"] = 1
```

```python
data_fake_manual_testing = data_fake
data_true_manual_testing = data_true
```

```python
data_fake_manual_testing.loc[:,"class"] = 0
data_true_manual_testing.loc[:,"class"] = 1
data_merge = pd.concat([data_fake, data_true],axis = 0)
data_merge.tail(20)
```

```python
data = data_merge.drop(['title','subject','date'],axis = 1)
```

```python
def wordopt(Text):
    text = Text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+','',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\n','',text)
    text = re.sub('\w*\d\w*','',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
#Defining Training and Testing Data and Splitting Them into &5 -25 Percent Ratio
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
#Converting Raw Data into Matrix for Further process

from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
#Creation of First Model

from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
#Checking the Model Accuracy

pred_lr = LR.predict(xv_test)
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
#Checking the Model Accuracy

pred_dt = DT.predict(xv_test)
DT.score(xv_test, y_test)
```
